### 讀入資料

```python
import pandas as pd
dataset = pd.read_csv(path+r'Social_Network_Ads.csv')
```

### 取出訓練特徵與標註

```python
X = dataset[['User ID', 'Gender', 'Age', 'EstimatedSalary']].values
Y = dataset['Purchased'].values
```

---

```python
import numpy as np
from sklearn.model_selection import KFold
```

### 將訓練資料按照順序切割成10等分

```python

kf = KFold(n_splits= 10, random_state= 43, shuffle=False)
kf.get_n_splits(X)

print(kf)
```

### 將訓練資料隨機切割成10等分

```python
kf = KFold(n_splits= 10, random_state= 43, shuffle=True)
kf.get_n_splits(X)

print(kf)
```

---

### 取出 切割資料對應位置

```python
train_split = kf.split(X)
next(train_split)
```

### Or

```python
for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
```

### 取出切割資料：trainset / testset 特徵(x_train/x_test)/標註(y_train/y_test)

```python
for train_index, test_index in kf.split(X):
  print(train_index, test_index)
  X_train, X_test = X[train_index], X[test_index]
  y_train, y_test = Y[train_index], Y[test_index]
```
